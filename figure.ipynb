```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure
fig, ax = plt.subplots(figsize=(10, 10))

# Set the aspect ratio of the plot to be equal
ax.set_aspect('equal')

# Draw concentric circles with larger radii
innermost_circle = plt.Circle((0, 0), 0.3, color='skyblue', fill=False, linewidth=2)
middle_circle = plt.Circle((0, 0), 0.6, color='lightgreen', fill=False, linewidth=2)
outer_circle = plt.Circle((0, 0), 0.9, color='lightcoral', fill=False, linewidth=2)

# Add the circles to the plot
ax.add_artist(innermost_circle)
ax.add_artist(middle_circle)
ax.add_artist(outer_circle)

# Add text to the center of each circle
plt.text(0, 0, "Computing-related impacts", horizontalalignment='center', fontsize=14, fontweight='bold')

# Subparts for Computing-related impacts
computing_subparts = ["ML inference", "ML training", "ML development"]
angles_inner = np.linspace(0, 2 * np.pi, len(computing_subparts), endpoint=False)

for i, label in enumerate(computing_subparts):
    x = 0.12 * np.cos(angles_inner[i])  # Adjusted for inner circle
    y = 0.12 * np.sin(angles_inner[i])  # Adjusted for inner circle
    plt.text(x, y, label, horizontalalignment='center', verticalalignment='center', fontsize=10)

# Add text to the middle circle (Application-related impacts)
middle_text = ["Buildings", "Power systems", "Energy supply", "Industry",
               "Transportation", "Agriculture", "Forestry and land use"]
angles_middle = np.linspace(0, 2 * np.pi, len(middle_text), endpoint=False)

for i, label in enumerate(middle_text):
    x = 0.45 * np.cos(angles_middle[i])  # Adjusted for middle circle radius
    y = 0.45 * np.sin(angles_middle[i])  # Adjusted for middle circle radius
    plt.text(x, y, label, horizontalalignment='center', verticalalignment='center', fontsize=10)

plt.text(0, 0.5, "Application-related impacts", horizontalalignment='center', fontsize=14, fontweight='bold')

# Add text to the outer circle (System-level impacts)
outer_text = ["Rebound effects", "Lock-in", "Path dependency"]
angles_outer = np.linspace(0, 2 * np.pi, len(outer_text), endpoint=False)

for i, label in enumerate(outer_text):
    x = 0.78 * np.cos(angles_outer[i])  # Outer circle radius
    y = 0.78 * np.sin(angles_outer[i])  # Outer circle radius
    plt.text(x, y, label, horizontalalignment='center', verticalalignment='center', fontsize=10)

plt.text(0, 0.85, "System-level impacts", horizontalalignment='center', fontsize=14, fontweight='bold')

# Remove the axes
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)

# Show the plot
plt.show()
```
